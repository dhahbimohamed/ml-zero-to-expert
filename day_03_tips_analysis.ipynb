```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

print(df.head())  # First 5 rows
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
print(df.describe())
print(df["sex"].value_counts())
print(df["day"].unique())
```

```python
print(df.groupby("sex")["total_bill"].mean())
```

```python
import matplotlib.pyplot as plt
plt.hist(df["total_bill"],bins=20,color="purple")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("total bill distribution")
plt.grid(True)
plt.show()
```

```python
tip_by_day=df.groupby("day")["tip"].mean()
print(tip_by_day)
```

```python
plt.bar(tip_by_day.index,tip_by_day.values,color="green")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip")
plt.title("Average Tip by Day of the Week")
plt.grid(True)
plt.show()
```

```python
df["tip_percent"]=df["tip"]/df["total_bill"]
print(df[["total_bill", "tip", "tip_percent"]].head())
```

```python
df.groupby("day")["tip_percent"].mean().plot(kind = "bar",color = "coral")
plt.title("Average Tip Percentage by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip Percentage")
plt.show()
```
